```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors

%matplotlib inline

plt.style.use("dark_background")
cmap = "viridis"
pi_chr = chr(960)
```

```python
x_max = 4
y_max = 3
x, y = np.meshgrid(
    np.linspace(-x_max, x_max, 4*x_max+1), 
    np.linspace(-y_max, y_max, 4*y_max+1)
)

# mag = np.hypot(1, dy)
u = 1/np.sqrt(2)
v = u
plt.quiver(
    x, y, 
    u, v, 
    1, 
    cmap = cmap,
    norm = colors.Normalize(vmin=0, vmax=2),
    pivot = "mid",
    scale = 16,
)
plt.colorbar()
plt.title("dy/dx = 1")
plt.tight_layout()
# plt.savefig(f"./svg/constant_slope_field_{cmap}.svg", dpi=600)

plt.show()
```

```python
x_max = 4
y_max = 3
x, y = np.meshgrid(
    np.linspace(-x_max, x_max, 4*x_max+1), 
    np.linspace(-y_max, y_max, 4*y_max+1)
)

mag = np.hypot(1, x)
u = 1/mag
v = x/mag
plt.quiver(
    x, y, 
    u, v, 
    np.abs(x), 
    cmap = cmap,
    pivot = "mid",
    scale = 19,
)
plt.colorbar()
plt.title("dy/dx = x")
plt.tight_layout()
# plt.savefig(f"./svg/parabola_slope_field_{cmap}.svg", dpi=600)

plt.show()
```

```python
x_max = 4
y_max = 3
x, y = np.meshgrid(
    np.linspace(-x_max, x_max, int(4*x_max+1)), 
    np.linspace(-y_max, y_max, int(4*y_max+1))
)
dy = np.exp(-y)
mag = np.hypot(1, dy)
plt.quiver(
    x, y, 
    1/mag, dy/mag, 
    dy, 
    cmap = cmap,
    norm = colors.LogNorm(vmin=1/32, vmax=32),
    pivot = "mid",
    scale = 19,
)
plt.colorbar()
plt.title("dy/dx = exp(-y)")
plt.tight_layout()
# plt.savefig(f"./svg/log_x_slope_field_{cmap}.svg", dpi=600)

plt.show()
```

```python
x_max = 4
y_max = 3
x, y = np.meshgrid(
    np.linspace(-x_max, x_max, int(6*x_max+1)), 
    np.linspace(-y_max, y_max, int(6*y_max+1))
)
d = np.vectorize(
    lambda x: (np.pi*x*np.cos(np.pi*x) - np.sin(np.pi*x))/(np.pi*x*x) if x else 0
)
dy = d(x)
mag = np.hypot(1, dy)
v = dy/mag
plt.quiver(
    x, y, 
    1/mag, v, 
    np.abs(dy), 
    cmap = cmap,
    pivot = "mid",
    scale = 27,
)
plt.colorbar()
plt.title(f"dy/dx = ({pi_chr}xcos({pi_chr}x)-sin({pi_chr}x))/({pi_chr}x^2)")
plt.tight_layout()
# plt.savefig(f"./svg/sinc_x_slope_field_{cmap}.svg")

plt.show()
```

```python
x_max = 4
y_max = 3
x, y = np.meshgrid(
    np.linspace(-x_max, x_max, 4*x_max+1), 
    np.linspace(-y_max, y_max, 4*y_max+1)
)
dy = -2*x*y
mag = np.hypot(1, dy)
u = 1/mag
v = dy/mag
plt.quiver(
    x, y, 
    u, v, 
    np.abs(dy), 
    cmap = cmap,
    pivot = "mid",
    scale = 20,
)
plt.colorbar()
plt.title("dy/dx = -2xy")
plt.tight_layout()
# plt.savefig(f"./svg/gaussian_slope_field_{cmap}.svg", dpi=600)

plt.show()
```

```python
x_max = 4
y_max = 3
x, y = np.meshgrid(
    np.linspace(-x_max, x_max, 4*x_max+1), 
    np.linspace(-y_max, y_max, 4*y_max+1)
)

dy = np.vectorize(lambda y, x:
    np.inf if x == 0 else y/x)(y, x)
mag = np.hypot(1, dy)
u = 1/mag
v = dy/mag
plt.quiver(
    x, y, 
    u, v, 
    np.abs(dy), 
    cmap = cmap,
    pivot = "mid",
    scale = 20,
)
plt.colorbar()
plt.title("dy/dx = y/x")
plt.tight_layout()
# plt.savefig(f"./svg/line_slope_field_{cmap}.svg", dpi=600)

plt.show()
```
